## while 循环
- 当一个条件保持真的时候while循环重复执行语句
- while 循环一定要有结束条件，否则很容易进入死循环
- while 循环的语法是：

    while loop-contunuation-conndition:

        Statement

```python
i=0
while i<10:
    print('hh')
    i += 1
```

## for 循环
- Python的for 循环通过一个序列中的每个值来进行迭代
- range(a,b,k)， a，b，k 必须为整数
- a: start
- b: end
- k: step
- 注意for 是循环一切可迭代对象，而不是只能使用range

## 嵌套循环
- 一个循环可以嵌套另一个循环
- 每次循环外层时，内层循环都会被刷新重新完成循环
- 也就是说，大循环执行一次，小循环会全部执行一次
- 注意：
> - 多层循环非常耗时
  - 最多使用3层循环

## 关键字 break 和 continue
- break 跳出循环，终止循环
- continue 跳出此次循环，继续执行

```python
for i in range(10):
    if i ==5:
        continue
    print(i)
```

```python
for i in range(10):
    for j in range(10):
        print(i,j)
```

```python
for a in range(1,10):
    for b in range(1,a+1):
        print(b,'x',a,'=',a*b,end=' ')
    print()
```

```python
for a in range(1,10):
    for b in range(1,10):
        Y=a*b
        print(a,'x',b,'=',Y,'',end="")
    print()
```

```python
for b in range(2,50):
    for a in range(2,b):
        if b%a==0:
            break
    else:
        print(b)
```

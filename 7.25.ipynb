```python
def isConsecutiveFour(lst = []):
    count_ = 1
    for i in range(0,len(lst)-1):
        if lst[i]==lst[i+1]:
            count_ += 1
    if count_ != 4:
        print('这个序列不是具有相同值的四个连续数字')
    if count_ == 4:
        print('这个序列是具有相同值的四个连续数字')
```

```python
isConsecutiveFour(lst = [1,5,1,1,1,1])
```

```python
isConsecutiveFour(lst=[1,2,5,6])
```
